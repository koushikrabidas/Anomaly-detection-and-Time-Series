1. What is Dimensionality Reduction? Why is it important in machine 
learning?
    - Dimensionality Reduction is the process of reducing the number of input variables or features in a dataset while preserving as much relevant information as possible. It transforms high-dimensional data into a lower-dimensional space.

Importance in ML:

Reduces computational cost and training time.

Helps remove noise and irrelevant features, improving model accuracy.

Prevents overfitting by simplifying the model.

Makes visualization easier for high-dimensional data.

4. Explain the concept of anomaly detection and its significance.
    - Anomaly detection refers to the identification of items, events, or observations that deviate significantly from the expected pattern of data.

Significance:

Helps detect fraud, intrusions, faults, or unusual events.

Prevents financial losses and improves security.

Essential for monitoring systems, networks, manufacturing lines, and health data

5. List and briefly describe three types of anomaly detection techniques.
    - Statistical Methods:
Identify anomalies by modelling normal behaviour using statistical distributions (e.g., Z-score, Gaussian models).

Machine Learning Methods:
Algorithms like Isolation Forest, One-Class SVM detect anomalies based on learned patterns.

Distance/Clustering-based Methods:
Anomalies are detected based on distance from clusters (e.g., k-NN, DBSCAN).

6. What is time series analysis? Mention two key components of time series

Key Components:

Trend: Long-term direction of data (upward, downward).

7. Describe the difference between seasonality and cyclic behavior in time 
series.
    - Seasonality:
Patterns that repeat at regular and fixed intervals (e.g., daily, monthly, yearly). Duration is predictable.

Cyclic Behavior:
Patterns that occur at irregular intervals depending on external factors (e.g., economic cycles). Duration is unpredictable.

8. Write Python code to perform K-means clustering on a sample dataset.

```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, y = make_blobs(n_samples=300, centers=3, random_state=42)

kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_

plt.scatter(X[:, 0], X[:, 1], c=labels)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200)
plt.title("K-Means Clustering")
plt.show()
```

9. What is inheritance in OOP? Provide a simple example in Python.

```python
class Animal:
    def speak(self):
        return "Animal makes a sound"

class Dog(Animal):
    def speak(self):
        return "Dog barks"

d = Dog()
print(d.speak())
```
